**Question No. 06**

***Write a programming construct (create a user defined function) to calculate the median value of the California Housing Data Set wherever required.***

```python
import pandas as pd

def calculate_median(feature_name):
    """
    This function reads the California Housing dataset and returns the median value of a given feature.
    """
    # Load the dataset into a pandas DataFrame
    housing_data = pd.read_csv('california_housing_data.csv')

    # Extract the feature column from the DataFrame
    feature = housing_data[feature_name]

    # Calculate the median of the feature
    median = feature.median()

    return median
```

```python
median_house_value = calculate_median('median_house_value')
print("The median house value is:", median_house_value)
```

*Using this we can calculate the median value for the California Housing Data whereever required.*
